```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Preprocessing

```python
data_df = pd.read_csv("Students_Data.csv")
```

```python
print(data_df.isnull().sum())
# counts number of nulls per column
# so we can see if there is any missing values
```

```python
print(data_df.isnull().sum() / len(data_df) * 100)
# counts percentage of nulls per column
```

```python
# If the header is still not parsed correctly, let's manually split the column names
if data_df.shape[1] == 1:
    # Get the column names by splitting the first row
    with open("Students_Data.csv", 'r') as f:
        first_line = f.readline().strip()
        column_names = first_line.split(';')
    
    # Now read the data without header and assign the column names
    data_df = pd.read_csv("Students_Data.csv", sep=';', header=None, skiprows=1)
    data_df.columns = column_names

# Check the column names again
print("\nColumn names:")
print(data_df.columns.tolist())
```

```python
# Visualization for Q1: Prior Academic Performance
# Comparing Admission grade vs Target

plt.figure(figsize=(10, 6))
sns.boxplot(x='Target', y='Admission grade', data=data_df)
plt.title('Distribution of Admission Grades by Student Outcome')
plt.xlabel('Student Outcome')
plt.ylabel('Admission Grade')
plt.savefig('Images\admission_grade_distribution.png')
plt.show()
```

```python
# Visualization for Q2: First-Semester Engagement
# Comparing Curricular units approved in 1st semester vs Target
plt.figure(figsize=(10, 6))
sns.boxplot(x='Target', y='Curricular units 1st sem (approved)', data=data_df)
plt.title('1st Semester Units Approved by Student Outcome')
plt.savefig('Images\first_semester_units.png')
plt.show()
```

```python
# Visualization for Q3: Socioeconomic Factors
# Scholarship holder status vs Target
plt.figure(figsize=(10, 6))
sns.countplot(x='Scholarship holder', hue='Target', data=data_df)
plt.title('Student Outcomes by Scholarship Status')
plt.xlabel('Scholarship Holder (0=No, 1=Yes)')
plt.ylabel('Count')
plt.savefig('Images\scholarship_status.png')
plt.show()
```
